```python
x = ['a','b','c','d','e']
```

```python
y = np.random.rand(5)
```

```python
plt.bar(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('bar plot')
```

```python
y
```

```python
plt.barh(x,y) #horizontal bar plot
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('bar plot')
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y,'--b')
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.grid() #for making boxes in the background
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y, c = 'brown', s= sizes, alpha = .5, marker = 'v')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('random data')
```
